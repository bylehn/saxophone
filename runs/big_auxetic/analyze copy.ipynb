```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_list = os.listdir('./')
results = []

for file_name in file_list:
    if file_name.endswith('.npz'):
        file = np.load(file_name)
        results.append([file['exit_flag'], file['poisson'], int(file_name[0])])

results = np.array(results)
print(results)
```

```python
print(file_list)
```

```python
plt.scatter(results[:,6],results[:,4], c=results[:,0], marker='o', cmap='jet', vmin=0, vmax=1)
#plt.xlim(-1,1)
plt.xlabel('poisson ratio')
plt.ylabel('Fraction of forbidden states')
plt.colorbar()
#plt.gca().set_aspect(10, adjustable='box')
#plt.legend()
#plt.savefig('results_forbidden_states_fraction.jpg')
plt.show()
```

```python
plt.scatter(results[:,4], results[:,2], marker='.', label='Initial')
plt.scatter(results[:,4], results[:,3], marker='.', label='Final')

plt.xlabel('poisson ratio')
plt.ylabel('Number of forbidden states')
plt.legend()
#plt.gca().set_aspect(10, adjustable='box')

plt.savefig('results_forbidden_states_fraction.jpg')
plt.show()
```

```python
plt.scatter(results[:,5],results[:,4], c=results[:,2], marker='.')
#plt.ylim(0,1)
plt.xlim(-0.55,0)
plt.ylim(-0.55,0)
plt.xlabel('Poisson ratio target')
plt.ylabel('Poisson ratio achieved')
plt.gca().set_aspect(1, adjustable='box')

plt.savefig('results_forbidden_states_fraction.jpg')
plt.show()
```

```python
plt.scatter(results[:,0], results[:,2], cmap = 'jet', marker='.')
plt.ylim(-0.5 , 3.5)
plt.xlabel('pertubation')
plt.ylabel('exit flag')
plt.gca().set_aspect(0.7, adjustable='box')

plt.savefig('exit_flags.jpg')
plt.show()
```
